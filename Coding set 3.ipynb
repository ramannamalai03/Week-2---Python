```python
#1. Square of Numbers from 1 to 5
Output:
1 4 9 16 25
```

```python
squares=[1,2,3,4,5]
for num in squares:
    square=num*num
    print(square, end=" ")
```

```python
 #3.Find Even and Odd Numbers in a List
Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].

```

```python
list=[12,17,19,24,35,40,56]
for num in list:
    if((num%2)==0):
        print("even numbers from the list:",num)
    else:
        print("odd numbers from the list:",num)
    
```

```python
#4. Reverse a Number
Question: Write a Python program to reverse the number 12345.

```

```python
num=12345
reverse=int(str(num)[::-1])
print(reverse)
```
